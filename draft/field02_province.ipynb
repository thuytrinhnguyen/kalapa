```python
import pandas as pd
```

### Load dataset

```python
data = pd.read_csv('train.csv', index_col=0)
```

```python
data.head(3)
```

### Field Province

```python
pd.isna(data.province).sum()
```

```python
data.province = data.province.str.lower()
```

```python
data.province.value_counts()
```

```python
data.province = data.province.replace('tỉnh hoà bình', 'tỉnh hòa bình')
```

```python
data.province.value_counts()
```

```python
data_groupby_province_sum_label = data.groupby(['province']).sum()
data_groupby_province_sum_label = pd.DataFrame(data_groupby_province_sum_label['label'])
data_groupby_province_sum_label.columns = ['bad_count']
data_groupby_province_sum_label
```

```python
data_groupby_province = data.groupby(['province']).count()
data_groupby_province = pd.DataFrame(data_groupby_province['label'])
data_groupby_province
```

```python
data_groupby_province_sum_label = data_groupby_province_sum_label.join(data_groupby_province)
```

```python
data_groupby_province_sum_label['bad_ratio'] = data_groupby_province_sum_label.bad_count / data_groupby_province_sum_label.label
```

```python
data_groupby_province_sum_label.sort_values('bad_count',ascending=False)
```
