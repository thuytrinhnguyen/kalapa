```python
import pandas as pd
```

### Load dataset

```python
data = pd.read_csv('train.csv', index_col=0)
```

### 1. Check nan value

```python
pd.isna(data.district).sum()
```

### 2. Lowercase for Categorical data

```python
data.district = data.district.str.lower()
data.province = data.province.str.lower()
```

### 3. Count unique values

```python
data.district.unique()
```

```python
data.district.value_counts()
```

### 4. Association with Bad Credit

```python
data_groupby_district_sum_label = data.groupby(['province','district']).sum()
data_groupby_district_sum_label = pd.DataFrame(data_groupby_district_sum_label['label'])
data_groupby_district_sum_label.columns = ['bad_count']
```

```python
data_groupby_district_sum_label
```

```python
data_groupby_district = data.groupby(['province','district']).count()
data_groupby_district = pd.DataFrame(data_groupby_district['label'])
data_groupby_district
```

```python
data_groupby_district_sum_label = data_groupby_district_sum_label.join(data_groupby_district)
```

```python
data_groupby_district_sum_label['bad_ratio'] = data_groupby_district_sum_label.bad_count / data_groupby_district_sum_label.label
```

```python
data_groupby_district_sum_label.sort_values('bad_ratio',ascending=False)
```
