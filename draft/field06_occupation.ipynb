### 4. Association with Bad Credit

```python
data_groupby_occupation_sum_label = data.groupby(['maCv']).sum()
data_groupby_occupation_sum_label = pd.DataFrame(data_groupby_occupation_sum_label['label'])
data_groupby_occupation_sum_label.columns = ['bad_count']
data_groupby_occupation_sum_label
```

```python
data_groupby_occupation = data.groupby(['maCv']).count()
data_groupby_occupation = pd.DataFrame(data_groupby_occupation['label'])
data_groupby_occupation
```

```python
data_groupby_occupation = data_groupby_occupation.join(data_groupby_occupation_sum_label)
data_groupby_occupation['bad_ratio'] = data_groupby_occupation.bad_count / data_groupby_occupation.label
```

```python
data_groupby_occupation.sort_values('bad_ratio', ascending=False)
```
